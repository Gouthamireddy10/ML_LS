```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
```

```python
# Load training data
data = pd.read_excel(r"C:\Users\tudig\Downloads\Training data.xlsx")
x_train = np.array(data.iloc[:, 0:8])
y_train = np.array(data.iloc[:, 8]).reshape(-1, 1)
```

```python
# Load test data
test_data = pd.read_excel(r"C:\Users\tudig\Downloads\Test data.xlsx")
x_test = np.array(test_data.iloc[:, 0:8])
y_test = np.array(test_data.iloc[:, 8]).reshape(-1, 1)
```

```python
# Plotting y_train against each feature in x_train
num_features = x_train.shape[1]
fig, axes = plt.subplots(num_features, 1, figsize=(10, 4*num_features))

for i in range(num_features):
    axes[i].scatter(x_train[:, i], y_train, marker='o', color='blue', alpha=0.6)
    axes[i].set_xlabel(f'Feature {i+1}')
    axes[i].set_ylabel('y_train')

plt.tight_layout()
plt.show()
```

```python
# Perform label encoding for categorical features if any
def feature_changing(x_train):
    labelencoder = LabelEncoder()
    for i in range(x_train.shape[1]):
        if x_train[:, i].dtype == object:
            x_train[:, i] = labelencoder.fit_transform(x_train[:, i])
    return x_train
x_train = feature_changing(x_train)
x_test = feature_changing(x_test)
```

```python
# Feature scaling using Z-score normalization
def z_score(x_train):
    x_mean = np.mean(x_train, axis=0)
    x_std = np.std(x_train, axis=0)
    x_train = (x_train - x_mean) / x_std
    return x_train, x_std, x_mean

x_train = x_train.astype(np.float64)
x_train, x_std, x_mean = z_score(x_train)
```

```python
# Define cost function (Mean Squared Error)
def cost(x_train, y_train, w, b):
    m = len(y_train)
    predictions = np.dot(x_train, w) + b
    loss = (1 / (2 * m)) * np.sum((predictions - y_train) ** 2)
    return loss
```

```python
# Define gradient descent function
def gradient_descent(x_train, y_train, w, b, learning_rate=0.01, num_iterations=1000):
    m = len(y_train)
    for i in range(num_iterations):
        predictions = np.dot(x_train, w) + b
        dw = (1 / m) * np.dot(x_train.T, (predictions - y_train))
        db = (1 / m) * np.sum(predictions - y_train)
        w -= learning_rate * dw
        b -= learning_rate * db
    return w, b

# Initialize weights
np.random.seed(2147483647)
w = np.random.randn(x_train.shape[1], 1)
b = np.random.randn(1)
```

```python
# Train the model using gradient descent
old_cost = 0
while abs(old_cost - cost(x_train, y_train, w, b)) > 0.00001:
    old_cost = cost(x_train, y_train, w, b)
    w, b = gradient_descent(x_train, y_train, w, b)

# Prepare test data
x_test = (x_test - x_mean) / x_std

# Make predictions on the test data
y_predict = np.dot(x_test, w) + b
```

```python
# Calculate accuracy
accuracy = 0
for dim in range(len(y_test)):
    if abs(y_predict[dim] - y_test[dim]) < 0.5:
        accuracy += 1
accuracy = round(accuracy * 100 / len(y_test), 2)
ok = 'Congratulations' if accuracy > 95 else 'Optimization required'
print(f"{ok}, your accuracy is {accuracy}%")
```
